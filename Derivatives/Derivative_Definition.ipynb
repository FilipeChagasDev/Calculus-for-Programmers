# Derivative (Differentiation) Definition

## Concept
The derivative of a real function $F(x)$ is a function $f(x)$ that maps the point rate of change from $F(x)$ for any $x$ value.

Geometrically, it can be said that $f(x)$ is the slope of a line that intersects the curve of $F(x)$ at two points $[x, F(x)]$ and $[x + h, F(x + h)]$ whose horizontal distance $h$ tends to zero.

![Secant Line example](https://upload.wikimedia.org/wikipedia/commons/9/97/Introductory_Physics_fig_1.15.png)

The derivative definition is:

$$f(x) = \lim_{h \rightarrow 0} \frac{F(x+h)-F(x)}{h}$$

## Notations

There are a few different notations for derivatives. See some of them below.

### Lagrange's notation
In this notation, the derivative of $F(x)$ is represented with a superscript $'$.

$$F'(x) = \lim_{h \rightarrow 0} \frac{F(x+h)-F(x)}{h}$$

Likewise, the second derivative (derived from the derivative) of $F(x)$ is represented as $F''(x)$, and the third derivative (derived from the derivative of the derivative) of $F(x)$ is represented as $F'''(x)$.

An alternative and more general form of this notation, using Arabic numerals, is to represent the nth derivative of $F(x)$ as $F^{(n)}(x)$.

### Leibniz's notation

This notation is more complex, but better represents the idea of the rate of change or slope of the derivative. In it, the derivative of $F(x)$ is represented as a fraction of the infinitesimal variations $dF$ and $dx$.


$$\frac{dF}{dx} = \lim_{h \rightarrow 0} \frac{F(x+h)-F(x)}{h}$$

Likewise, the nth derivative of $F(x)$ is represented as $\frac{d^n F}{dx^n}$.

## Numerical Differentiation

From [Wikipedia](https://en.wikipedia.org/wiki/Numerical_differentiation): *In numerical analysis, numerical differentiation describes algorithms for estimating the derivative of a mathematical function or function subroutine using values of the function and perhaps other knowledge about the function. \[...\] The simplest method is to use finite difference approximations.*

Let's make a numerical differentiation of $y(x) = x^3$ by finite difference approximation. This method basically consists of applying the formula $\frac{y(x+h)-y(x)}{h}$ to a value of $h$ small enough. For this, we will use the **NumPy** and **MatPlotLib** libraries.

First, we must import the packages:

```python
import numpy as np
from matplotlib import pyplot as plt
```

Now, let's create a vector with the values of $x$ and another with the values of $y(x)$. For $x$, we will use 100 values from -10 to 10.

```python
x = np.linspace(-10,10,100) #create a vector with 100 values from -10 to 10.
y = x**3 #apply y(x) for all x's values
```

Let's plot $y(x)$.

```python
plt.plot(x,y)
plt.show()
```

Now we are going to apply numerical differentiation using the following formula:
$$y'_i = \frac{y_{i+1}-y_{i}}{x_{i+1}-x_{i}}$$
Note that this formula is equivalent to $\frac{y(x+h)-y(x)}{h}$ considering $h=x_{i+1}-x_i$.

```python
ydiff = [(y[i+1]-y[i])/(x[i+1]-x[i]) for i in range(x.shape[0]-1)] #Apply the formula of y' for all y(x) values
ydiff = np.array(ydiff) #Convert the ydiff list to a numpy array
```

Now let's plot $y'$.

Note that **ydiff** has a value less than $y$ and $x$. Therefore, we need to use the slice *\[0:99\]* (99 values after index 0) of $x$.

```python
plt.plot(x[0:99],ydiff)
plt.show()
```

The symbolic derivative of $x^3$ is $3x^2$. Let's plot $3x^2$ to see if it looks like **ydiff**.

```python
fig, ax = plt.subplots()
ax.plot(x, 3*x**2, label='Symbolic')
ax.plot(x[0:99], ydiff, label='Numerical')
leg = ax.legend();
```
